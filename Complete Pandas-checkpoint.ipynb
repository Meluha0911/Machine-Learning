```python
# A pandas series is like a column in a table or it is one dimensional array which hold data of any type. here we will 
# create a simple pandas series
# In pandas, a Series is a one-dimensional labeled array that can hold various data types such as numbers, strings, dates, 
# or even more complex objects. It's like a single column of data in a table or a spreadsheet.
# Here are the key characteristics of a pandas Series:
# 1.One-Dimensional: A Series is essentially a single list of values, and it has a single dimension. You can think of it as 
#                    a column in a table.
# 2.Labeled Data: Each element in a Series has a label or an index associated with it. By default, pandas assigns an integer
#                 index starting from 0, but you can specify custom labels if needed. These labels help you access and 
#                 manipulate the data efficiently.

# 3.Homogeneous Data: Unlike Python lists, which can hold mixed data types, a Series typically holds data of the same type. 
#                     For example, you can have a Series of integers, a Series of strings, etc.
sanket = [1,7,2]
sanketnew = pd.Series(sanket)
sanketnew
```

```python
# A DataFrame in pandas is like a table of data, similar to what you might see in a spreadsheet or a database. It's a 
# two-dimensional data structure that consists of rows and columns, where each column can contain different types of 
# information like numbers, text, or dates.
# Here's a simple way to think about a pandas DataFrame:

# Rows: Each row in a DataFrame represents a separate entry or record, like a single person's information in a list of 
#       people, or a single day's data in a list of daily temperatures.
# Columns: Each column in a DataFrame represents a specific type of information, like a person's name, age, or address in a 
#          list of people, or different types of data like temperature, humidity, and date in a weather dataset.
# So, in essence, a DataFrame is a way to organize and work with data in a structured and tabular form, making it easy to 
# analyze, filter, and manipulate the information as needed for various data-related tasks. It's a powerful tool for data 
# analysis in Python.
```

```python
# In pandas, both `loc` and `iloc` are methods used for selecting data from a DataFrame, but they differ in how they perform
# the selection and what types of labels or indices they use:

# 1. - `loc` (Label-Based Selection):
#    - `loc` is primarily label-based, which means you use it to select data using row and column labels.
#    - You specify row and column labels explicitly, and it includes the endpoint.
#    - The syntax for `loc` is `df.loc[row_label, column_label]`, where `df` is your DataFrame.
#    - You can use label slices, boolean indexing, and callable functions with `loc`.label means any label 0,1,2,a,b,c etc.
#    - Example:
    
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data, index=['X', 'Y', 'Z'])
# Using loc to select data by labels
selection = df.loc['Y', 'A']
selection
```

```python
# 2. - `iloc` (Integer-Based Selection):
#    - `iloc` is primarily integer-based, which means you use it to select data using integer positions (0-based indices).
#    - You specify row and column indices explicitly, and it includes the start index but excludes the end index.
#    - The syntax for `iloc` is `df.iloc[row_index, column_index]`, where `df` is your DataFrame.
#    - You cannot use label-based selections or slices with `iloc`. Here dont cosider about label just think about indexes.
#    - Example:
    
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
# Using iloc to select data by integer positions
selection = df.iloc[1, 0]
selection
```

```python
# Here's a summary of the key differences:

# - `loc` uses labels, while `iloc` uses integer positions.
# - `loc` includes the endpoint in slices, whereas `iloc` excludes it.
# - `loc` allows label-based selection, boolean indexing, and callable functions, making it more versatile for selecting 
#    data based on labels.
# - `iloc` is limited to integer-based indexing, which is useful for selecting data by position but doesn't work with labels
# - In most cases, you can choose between `loc` and `iloc` based on whether you want to select data by labels or integer 
#   positions, depending on your specific data analysis needs.
```

```python
# The agg() method in Pandas is a powerful and flexible way to apply one or more aggregation functions to one or more 
# columns in a DataFrame. It allows you to compute multiple summary statistics simultaneously or even apply custom
# aggregation functions. The agg() method can be used at both the DataFrame level and the group level when using groupby.
```

```python
# DataFrame Level Aggregation:
# You can use agg() to compute multiple aggregation functions for one or more columns in a DataFrame. For example:
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 22, 35, 28],
    'Salary': [50000, 60000, 45000, 75000, 52000],
    'Department': ['HR', 'Finance', 'IT', 'Finance', 'IT']
}

df = pd.DataFrame(data)

# Compute multiple aggregation functions for the 'Age' and 'Salary' columns
agg_result = df.agg({'Age': ['min', 'max', 'mean'], 'Salary': ['min', 'max', 'mean']})

print(agg_result)
```

```python
# Group Level Aggregation using groupby:
# You can also use agg() in combination with groupby to perform aggregation within groups. For example, calculating multiple
# aggregation functions for each department:
# Calculate multiple aggregation functions within each department group
department_agg = df.groupby('Department').agg({'Age': ['mean', 'std'], 'Salary': 'max'})
department_agg
# In this example, we calculate the mean and standard deviation of 'Age' and the maximum 'Salary' within each department.
```

```python
# You can use a dictionary to specify which aggregation functions to apply to each column. The keys of the dictionary are 
# column names, and the values are lists of aggregation functions or a single aggregation function. The result is a 
# DataFrame with hierarchical columns indicating the column and aggregation function applied.
# Additionally, you can use custom aggregation functions by defining your own Python functions and passing them to agg().
# This provides great flexibility when working with complex data analysis tasks.
# In summary, the agg() method in Pandas is a versatile tool for summarizing and aggregating data in DataFrames, whether 
# you're working with the entire DataFrame or performing group-level aggregations.
```

# Apply Method 

```python
# The apply method in Pandas is a powerful and flexible function that allows you to apply a custom function to each element
# or row of a DataFrame or Series. It's a way to transform or manipulate your data according to your specific needs. Here's
# a breakdown of how the apply method works:
```

```python
# Applying a Function to a Series:
# You can use apply on a Pandas Series to apply a function to each element of that Series. For example, if you have a Series
# of numbers and you want to square each number:
```

```python
data = {'Numbers': [1, 2, 3, 4, 5]}
series = pd.Series(data['Numbers'])
series
```

```python
def square(n):
    return n**2
```

```python
new_series = series.apply(square)
new_series
# Here, the square function is applied to each element in the 'Numbers' Series.
```

```python
# Applying a Function to a DataFrame:
# You can also use apply on a DataFrame to apply a function to either rows or columns of the DataFrame. By specifying the
# axis parameter, you can control whether the function is applied to rows (axis=0) or columns (axis=1). For example, if you
# have a DataFrame of numbers and you want to find the sum of each row:
```

```python
data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
df = pd.DataFrame(data)
df
```

```python
def row_sum(n):
    return n.sum()
```

```python
rowsum = df.apply(row_sum,axis=0)
rowsum
# Here, the row_sum function is applied to each row (axis=1) to calculate the sum of values in each row.
```

```python
df = pd.DataFrame(rowsum)
df
```

```python
# Applying Lambda Functions:
# You can also use lambda functions with apply for more concise operations. For example, if you want to calculate the 
# square root of each element in a Series:
```

```python
data = {'Numbers': [1, 4, 9, 16, 25]}
series = pd.Series(data['Numbers'])
series
```

```python
new_series = series.apply(lambda x : x**2)
new_series
# Here, a lambda function is used to calculate the square root of each element in the 'Numbers' Series.

# The apply method is incredibly useful when you need to perform custom operations on your data, and it allows you to work
# with your data in a more flexible and tailored way. It's a key tool in data manipulation and analysis using Pandas.
```

```python
# map is used to apply a function to each element of a Series. It's like saying, "For each item in this list, do something."
# apply is used to apply a function to each row or column of a DataFrame. It's like saying, "For each row (or column) in 
#this table, do something."
# So, map is for working with individual items in a Series, while apply is for working with rows or columns in a DataFrame.
```

```python
# The apply and map functions in Pandas are used to transform data in DataFrames and Series by applying custom functions. 
# However, they have different use cases:
```

```python
# 1.apply Function:
# Use: apply is used to apply a function along the rows or columns of a DataFrame or to each element of a Series.
# Example (DataFrame): You can use apply to perform operations on entire rows or columns of a DataFrame. For instance, you 
# can calculate the sum of values in each row or the mean of values in each column.
df.apply(sum, axis=1)  # Apply the sum function to each row
df.apply(mean, axis=0)  # Apply the mean function to each column
#Example (Series): When used with a Series, apply allows you to apply a function to each element of the Series.
series.apply(lambda x: x * 2)  # Double each element in the Series
```

```python
# 2.map Function:
# Use: map is primarily used with Series and is designed for element-wise transformations.
# Example (Series): You can use map to apply a function to each element in a Series, creating a new Series with the 
#                   transformed values.
series.map(lambda x: x + 10)  # Add 10 to each element in the Series
# Example (DataFrame): While less common, you can use map with DataFrames to transform specific columns (Series) by 
#                      providing a dictionary mapping the column to a function.
df['Column1'] = df['Column1'].map(some_function)
```

```python
# Use apply when you want to perform operations that involve multiple elements in a row or column of a DataFrame, or when 
# you want to apply a function to each element in a Series.
# Use map when you want to apply a function element-wise to a Series, effectively transforming each value individually.
# Both functions are essential tools for data manipulation in Pandas, but their specific use cases differ based on the type
# of data structure (DataFrame or Series) you are working with and the nature of the transformation you want to apply.
```

# Append Function

```python
# Certainly! In simple terms, the Pandas append function is like adding one table (DataFrame) to the bottom of another 
# table.It combines data from two tables vertically, stacking one on top of the other.
# Imagine you have two tables with the same columns, one containing information about students in a class and another
# containing information about students in another class. If you want to combine both classes into a single table, you can
# use append. It's as if you're stacking the second table below the first one, creating a larger table with all the 
# students' data.
# Here's a simplified example:
import pandas as pd

# Create the first DataFrame
data1 = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df1 = pd.DataFrame(data1)

# Create the second DataFrame
data2 = {'Name': ['Charlie', 'David'], 'Age': [22, 35]}
df2 = pd.DataFrame(data2)

# Append the second DataFrame to the first one
combined_df = df1.append(df2, ignore_index=True)

combined_df

# In this example, the append function combines df2 with df1, creating combined_df, which contains all the student data
# from both classes. The ignore_index=True argument ensures that a new continuous index is created for the combined 
# DataFrame.
```

```python
# Certainly! In simple terms, the Pandas merge function is like combining two tables (DataFrames) by matching common 
# information between them, just like merging pieces of a jigsaw puzzle.
# Imagine you have two tables, one containing information about students and another containing information about their
# test scores. The merge function helps you bring these two tables together based on a common column, such as a student ID,
# to create a new table that combines both pieces of information into one.
```

```python
# Here's a simplified example:
# Create the first DataFrame with student information
data1 = {'StudentID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
student_df = pd.DataFrame(data1)
student_df
```

```python
# Create the second DataFrame with test scores
data2 = {'StudentID': [2, 3, 4], 'Score': [85, 92, 78]}
scores_df = pd.DataFrame(data2)
scores_df
```

```python
# Merge the two DataFrames based on the 'StudentID' column
merged_df = pd.merge(student_df, scores_df, on='StudentID')
merged_df
```

```python
# In this example, the merge function combines student_df and scores_df based on the 'StudentID' column. It matches the 
# student records from both tables and creates a new DataFrame (merged_df) that includes both the students' names and their
# test scores.
# So, in simple terms, Pandas merge helps you put together related information from different tables by finding common 
# elements (like the student ID in this case) and creating a single, combined table. It's like connecting the dots between
# two sets of data.
```
