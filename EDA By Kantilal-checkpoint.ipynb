# Detection & Removal of Outlier's.

```python
# Today we will look how to detect outlier's and remove them.
# Feature main outlier's hai ya nahi wo hum kiase dekhange -- scatterplot, boxplot.
# Today we will see how to remove outlier's.
# Firstof of all outlier ko remove karnekeliye we to see distribution. -- Ek feature ka distribution kaise hota hai like --
# jababhi hume kabhi aisa expected hai ki jo kuch features hai uska jo feature ka jo distribution hai that should be in 
# bell curved why -- bcz maximum data around the mean -- jo mean rahega uske around rehana chahiye. -- To aisa std rule 
# hai agar aapka feature bell curved main hai iska matlab -3-2-1 to +1+2+3 -- agar manlo medical ya financial domain pe hum 
# kam kar rahe hai jahapar risk factor jyada hai to we need to consider the concentrated data -- around mean.  Jitna scatter
# hoga -- outlier kya hota hai -- a data point which is far away from actual observation's -- means jo data humare obsrvtnke
# bahot door hai wo outlier hota hai. For eg. -3 and +3 ke above ka data. To standard Deviation it is a measurement -- 
# meanse kitna door hai iska measurement it is a standard deviation. Deviate kitna kar raha hai wo, vary kitna kar raha hai
# wo iska measurement. SD ka formula -- Har ek observationmese humne mean minus kiya -- jab hume koi difference nikalana
# hai -- for eg mujhe 2 and 8 ka diff nikalana hai to 8-2 = 6 to 2 ye 8 se 6 points door hai. SD yahape imp role play karega
# bcz usase hume pata chalega ki particular datapoint ye humare meanse kitna door hai -- doori mapnekeliye we can use SD.
# The will not ask you the formula of SD in the Interview.
# So -3 ke below or above +3 ke jo datapoints hai wo humareliye outliers hai.
# Outliers ko (nikalnekeliye) Detect & Remove humare paas do technique hai. Ek hota hai Z Score Technique aur dusra hota 
# hai IQR(Inter Quartile Range). Donobhi Technique ek jaise kam karte hai -- jaise bell curve shape hota hai usike tarikese
# boxplot ka IQR hota hai. (see SS) -- 0 percentile, 25th percentile(Q1), 50th percentile(Q2), 75th percentle(Q3),
# 100th percentile. -- To hum 25 to 75 yani first Q to Third Q tak ka data consider karte hai -- itnehi range ka data we 
# consider karte hai -- and itnehi range ko hum bolte hai IQR(Inter Quartile Range). so IQR is nothing but data from 
# Q1 to Q3 . So formula of IQR = Q3-Q1 so isase hume IQR vales milegi -- aue usi IQR ki value ko use karkar hum lower bound
# aur higher end nikalenge.  Lower Bound kaise nikalenge -- to LB = Q1-1.5*IQR.
```

```python
# Detecting & Removing Outlier's Using IQR Technique.
```

```python
# Formulas
# IQR = Q3 -Q1
# Lower Bound(LB) = Q1-1.5*IQR
# Upper Bound(UB) = Q3+1.5*IQR
```

```python
# Example
# Age = 16,15,14,13,15,16,14,17,19,-600,500,9000,460,16,17,18
# IQR main hume first isko arrange karna padega kyunki hume median nikalna hai. To pehale hum age ke data ko ascending order
# yani sort karlenge. -600,13,14,14,15,15,16,16,16,17,17,18,19,75,460,500,9000
# Ab yahase hume median nikalana hai -- so here median is 16
# Ab hume IQR value nikalni hai -- mujhe Q3 and Q1 nikalana hai -- means mujhe 25th percentile and 75th percentile nikalana
# hai -- to yahape 25ht percentile(Q1) aayega = 14 and 75th percentile(Q3) will be 18
# So IQR = 18-14 i.e 4
# Now LB = Q1-1.5*IQR i.e 14-1.5*4 = 8
# UB = Q3-1.5*IQR i.e 18+1.5*4 = 12
# Means humareliye jo kuch value 8 se kum hai aur jo kuch value24 se above hai wo sab humareliye outliers hai. Abhi mujhe 
# age chahiye without outliers. means mujhe sirf wahi age chahiye jo 8 aur 24 ke bichme ho.
# age = 16,15,14,13,15,16,14,17,19,16,17,18 -- to yahape -600,500,9000,460 ye outliers hai wo nikalgaye using IQR.
# IQR main below LB and above UB data is outlier.
```

```python
# Now we will check outliers in age -- ageme dekhaneliye jaise humne Quartiile nikala tha 25,75 ye fix values nahi hai 
# depending on the domain hum Quartile lete hai
q1 = df.Age.quantile(0.25)  # Syntax to get the Quartile 0.25 means Q1
```

```python
# Z score formula kya bolta hai ki agar aap harek datapoint - mean/SD to ye hume ek value dega .i.e zscore ka value. To us 
# zscore ka value hum use kar sakte hai to find out the lower_bound and the upper_bound.
# z_score = xi-mean/SD
# Agar manlo aappko 3th SD , 4th SD main jana hai
```

```python
df
```

```python
df['z_score'] = (df['Age']-df['Age'].mean())/df['Age'].std() # This will give us Z_Score.
```

```python
df
```

```python
# Ab z_score humara length define karega. Hume kase kahatak dekhana hai -- that means agar hume -3 to +3 in scenario pe 
# utnahi dataset aayega agar hum outlier nikalenge to.
df1 = df[(df['z_score']>-3) & (df['z_score']<3)] # To outlier exclude karke record dega ye hume
df1
```
