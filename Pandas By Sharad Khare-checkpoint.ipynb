# SERIES

```python
# A pandas series is like a column in a table or it is one dimensional array which hold data of any type. here we will create 
# a simple pandas series
sanket = [1,7,2]
sanketnew = pd.Series(sanket)
sanketnew
```

```python
# Accessing data of series that process is known as labeling - label can be used to access a specified value
sanketnew[1]
```

```python
# With create label's we can create our own name lables(index).
sanket = [1,7,2]
sanketnew = pd.Series(sanket,index = [1,2,3])
sanketnew
```

```python
sanket = [1,7,2]
sanketnew = pd.Series(sanket,index = ['x','y','z'])
sanketnew
#In short we change the indexing here as per our parameter's
```

```python
sanketnew['y']
```

```python
# You can use a key or value object like a dictionary , when creating a series. Here we will create series from a dictionary.
data = {'name':'sanket','age':24,'salary':198000}
da = pd.Series(data)
da
```

```python
da['name'] 
```

```python
da['salary']
```

```python
# Now we will create series only from name and salary i.e we will use only two key's of dictionary to create series
data = {'name':'sanket','age':24,'salary':198000}
da = pd.Series(data,index = ['name','salary'])
da
```

# DATAFRAME

```python
# Dataset's in pandas are usually multidimensional table's and they are called dataframe's.
#Series are like column's while Dataframe are like whole Table.
# DataFrame is a Two-Dimensional data structure eg. Table with Row's and Column's
```

```python
#Create dataframe from 2 Series
sanket = {'cal':[420,380,390],
         'duration':[50,40,45]}
sanketnew = pd.DataFrame(sanket)
sanketnew
```

```python
# DataFrame using dictionary
data = {'name':['sanket','shital','rohit','mayur'],
       'age':[24,28,32,34],
       'salary':[198000,78000,48000,42000],
       'city':['satara','satara','satara','mumbai']}
result = pd.DataFrame(data)
result
```

```python
# How to show particular row or column from dataframe -- using locaterow attribute -- pandas use loc attribute to return one 
# or more specified row's
result.loc[0]
```

```python
# Access multiple row's at a time using loc attribute from a dataframe
result.loc[[0,1]]
```

```python
result.loc[[0,1,3]]
```

```python
# With the index argument we can name our own index -- Named Index
data = {'name':['sanket','shital','rohit','mayur'],
       'age':[24,28,32,34],
       'salary':[198000,78000,48000,42000],
       'city':['satara','satara','satara','mumbai']}
result = pd.DataFrame(data,index = ['na','ag','sa','ci'])
result
#Alway's pass group of values in the form of list(array)
```

```python
#Accessing multiple value using loc method
result.loc[['na','ag']]
```
