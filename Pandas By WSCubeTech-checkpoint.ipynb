```python
# here we had use list to make series in pandas but we can also use dictionary, tuple etc,.
#convert dictionary data into series
dic = {'name':['python','c','c++','java'],
      'por':[12,13,14,15],
      'rank':[1,4,3,2]}
var = pd.Series(dic)
var
#here datatype is object bcz we use mix datatype here of string and int so whenever we use mixdatatype we will get object.
```

```python
# how create series by single data
s = pd.Series(12)
s
#series from single number
```

```python
# how to create multiple block of single data in a series --> for that you have to pass no of index values
var = pd.Series(12,index=[1,2,3,4,5])
var
#series from single number.
```

```python
# we use series over numpy array bcz in series there is no broadcasting rule
var1 = pd.Series(12,index=[1,2,3,4,5,6,7])
var2 = pd.Series(12,index=[1,2,3,4])
#here one series has 7 data while second series has 4 data --> we can perform add on of this two series -->here upto 4 data 
# is equal so data till 4 will be added but ahead of that you will get NaN but this is not possible in NumPy it will give 
# broadcast error
var1+var2
# series also work with missing data that is the benefit pandas has over NumPy
```

```python
var1 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['a','b','c','d'])
var2 = pd.DataFrame({'C':[10,20],'D':[11,22]},index=['a','b'])
var2.join(var1)
```

```python
var1 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['a','b','c','d'])
var2 = pd.DataFrame({'C':[10,20],'D':[11,22]},index=['a','b'])
var2.join(var1,how = 'left')
```

```python
var1 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['a','b','c','d'])
var2 = pd.DataFrame({'C':[10,20],'D':[11,22]},index=['a','b'])
var2.join(var1,how='right')
```

```python
var1 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['a','b','c','d'])
var2 = pd.DataFrame({'C':[10,20],'B':[11,22]},index=['a','b'])
var2.join(var1,how='outer',lsuffix='_12') # here the name of the left column change in the runtime
# suffix is use to change the column_name in the runtime in oerder to avoid error
```

```python
#APPEND 
# this function is also use to join two DF
# diff between join and append function is that in join function when column_name is same it thow's error so for that we
# have to use the suffix parameter this is not the case in the append function even if data is same
var1 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]},index=['a','b','c','d'])
var2 = pd.DataFrame({'C':[10,20],'B':[11,22]},index=['a','b'])
var1.append(var2)
```

```python
# GROUP BY FUNCTION IN THE PANDAS
# in pandas if you want to arrange random data in sequential manner then we use group by function in the pandas
```

```python
var = pd.DataFrame({'name':['a','b','c','d','a','b','a','b','a','c','c'],
                   's_1':[12,13,14,12,13,14,15,23,25,16,10],
                   's_2':[23,24,25,26,27,28,29,30,25,34,35]})
var
```

```python
# i can group by above data base on the name like a,a,a,a,a,b,b,b,b,b,c,c,c,c and so on
var_new = var.groupby('name') # mention how to group by that data
var_new #--> to show this data use for loop
```

```python
for x,y in var_new:
    print(x)
    print()
    print(y) #here data will be shown groupby along with the respective number's.
```

```python
# how to see single group's datacsay data of a group
var_new.get_group('a') # this will show data of group 'a'.
```

```python
#how to get min,max,mean from different group's
var_new.min() #by this we will get min data of a,b,c,d
```

```python
var_new.max() #this will give max values from a,b,c,d
```

```python
var_new.mean() #this will give mean values
```

```python
#how to convert group_by result into list
li = list(var_new)
li #after converting data into list you can perform multiple operation on that data.
```

```python
#MELT
#
var = pd.DataFrame({'day':[1,2,3,4,5,6],
                  'eng':[10,12,14,15,16,12],
                  'maths':[17,18,19,13,14,16]})
var #how to reshape this file
```

```python
pd.melt(var) # table become vertically from horizontal format
```

```python
#how to make any value as id
pd.melt(var,id_vars=['eng']) #english data work as a id here.
```

```python
pd.melt(var,id_vars=['day'])
```

```python
#how to customise name here variable 
pd.melt(var,id_vars=['eng'],var_name='python')
```

```python
#how to change name of value's
pd.melt(var,id_vars=['eng'],var_name='python',value_name='wscube')
```

```python
#PIVOT
#this function is use for data reshaping
var = pd.DataFrame({'day':[1,2,3,4,5,6],
                    'st_name':['a','b','c','a','b','c'],
                  'eng':[10,12,14,15,16,12],
                  'maths':[17,18,19,13,14,16]})
var
```

```python
# i want set data in the above table according to the student name for that we will use pivot
var.pivot(index= 'day',columns='st_name')  #here you have to give two argument's here i want to make days index and mention 
#column accordgto which my data has to be arranged
```

```python
#how to get single data say data of only english data getting particular column data
var.pivot(index= 'day',columns='st_name',values='eng')
```

```python
var = pd.DataFrame({'day':[1,1,1,1,2,2],
                    'st_name':['a','b','c','a','b','c'],
                  'eng':[10,12,14,15,16,12],
                  'maths':[17,18,19,13,14,16]})
#how to  arrangee in proper way
var.pivot_table(index='st_name',columns='day',aggfunc='mean')  #perform aggregate function
```

```python
var.pivot_table(index='st_name',columns='day',aggfunc='sum') 
```

```python
var.pivot_table(index='st_name',columns='day',aggfunc='sum',margins=True) #it gives you mean value 
```
